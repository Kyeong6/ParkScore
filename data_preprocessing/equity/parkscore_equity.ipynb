```python
# 읍/면/동 추출
def extract_eup_myeon_dong(address):
    # Regular expression pattern to match 읍/면/동, including cases like "읍내동"
    pattern = r'\w+(?:읍|면|동)'
    matches = re.findall(pattern, address)
    if matches:
        # Join matches to handle cases like "읍내동"
        return " ".join(matches)
    return None
```

## "구"로 나눈 데이터 Gini 계수 확인

- 해당 프로젝트에서의 "공평"이란 각 도시 별로 "읍/면/동"에 위치하는 도시공원의 배치가 "공간적 형평성"을 가지고 있는지 파악하는 지표
- 위에서 얻은 데이터셋(final_equity_gini.csv)을 수작업으로 수정 및 추가한 뒤 Gini 계수 적용

```python
# 데이터 타입 변환(,가 존재)
df['도시공원 면적'] = df['도시공원 면적'].str.replace(',', '').astype(float)
```

```python
# Gini 계수 적용
def calculate_gini(values):
    """Calculate the Gini coefficient using the Lorenz curve method."""
    values = np.sort(values)
    n = len(values)
    if n == 0 or np.all(values == 0):  # Handle edge cases with no values or all zero values
        return 0
    index = np.arange(1, n+1)
    return (np.sum((2 * index - n - 1) * values)) / (n * np.sum(values))
```

```python
# "도시공원 면적" 숫자형으로 변경
df['도시공원 면적'] = pd.to_numeric(df['도시공원 면적'], errors='coerce')
```

```python
# "도시공원 면적"에 존재하는 결측치 제거
df_cleaned = df.dropna(subset=['도시공원 면적'])
```

```python
# "도시공원 면적"은 제곱미터이므로 "도시면적"과 맞추기위한 작업 (1 square kilometer = 1,000,000 square meters
df_cleaned['도시공원 면적'] = df_cleaned['도시공원 면적'] / 1_000_000
```

```python
# 각 도시 내 행정 구역 수준의 공원 면적을 통합
aggregated_data = df_cleaned.groupby(['도시명', '행정구역'])['도시공원 면적'].sum().reset_index()
```

```python
# 각 도시의 행정구역 공원지역 분포에 대한 지니계수를 계산
gini_results = aggregated_data.groupby('도시명')['도시공원 면적'].apply(lambda x: calculate_gini(x.values))
```

```python
# DataFrame 변환
gini_df = gini_results.reset_index()
gini_df.columns = ['도시명', 'Gini계수']
```
